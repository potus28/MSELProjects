$q_{st} = -\frac{<UN>-<U><N>}{<N^2>-<N>^2} +RT$

```python
def qst(df):
    
    df["UN"] = df["Nmols_2"] * df["Energy_Total (kJ/mol)-Ext"]
    df["N^2"] = df["Nmols_2"] * df["Nmols_2"]
    
    numerator = np.mean(df["UN"]) - np.mean(df["Energy_Total (kJ/mol)-Ext"])*np.mean(df["Nmols_2"])
    denominator = np.var(df["Nmols_2"])
    
    q  = -1.0*numerator/denominator + R*300.0/1000.0
    
    return q
```
